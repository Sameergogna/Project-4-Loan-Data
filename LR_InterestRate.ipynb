### Outliers treament

```python
# Outlier capping Method
def outlier_capping(x):
    x = x.clip(upper = x.quantile(0.99))
    x = x.clip(lower = x.quantile(0.01))
    return x
```

```python
df_num=df1.drop(columns=["Loan.Purpose", "Home.Ownership","LoanID"])
df_cat=df1[["Loan.Purpose", "Home.Ownership"]]
```

```python
df_num1=df_num.apply(outlier_capping)
```

```python
df_cleaned=pd.concat([df_num1, df_cat], axis=1)
```
